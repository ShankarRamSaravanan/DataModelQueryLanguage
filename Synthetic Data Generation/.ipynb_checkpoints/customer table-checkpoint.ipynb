```python
List=[]
def First_Last(No_of_values):   
    for i in range(No_of_values):
        URL = 'https://randomuser.me/api/?nat=us'.format(API_VERSION)   
        JSON_FILE = json.loads(request.urlopen(URL).read())
        pattern = '[A-Z]+[a-z]'
        firstname = (JSON_FILE['results'][0]['name']['first'])
        lastname = (JSON_FILE['results'][0]['name']['last'])
        if re.search(pattern, firstname) and re.search(pattern, firstname):
            Title =(JSON_FILE['results'][0]['name']['first']) +" "+(JSON_FILE['results'][0]['name']['last'])+","+(JSON_FILE['results'][0]['email'])+','+(JSON_FILE['results'][0]['id']['value'])+','+JSON_FILE['results'][0]['phone']+','+str(JSON_FILE['results'][0]['location']['street']['number']) +' '+ JSON_FILE['results'][0]['location']['street']['name'] + ' '+ JSON_FILE['results'][0]['location']['city'] + ' '+  JSON_FILE['results'][0]['location']['state'] + ' '+  JSON_FILE['results'][0]['location']['country'] +' '+  str(JSON_FILE['results'][0]['location']['postcode'])
            
            List.append(Title)
        else :
            continue
        
start = time.time_ns()
values =10000
First_Last(values)    
while len(List)!= values:
    First_Last(values- len(List))
end = time.time_ns() 
print("Time taken", end-start, "ns")
```

```python
Customer =[] 
for i in range(len(List)):
    Customer.append(List[i].split(","))
Customer
```

```python
Name=[]
Email=[]
Phone_no=[]
SSN=[]
Address=[]
for i in range(len(Customer)):
    Name.append(Customer[i][0])
    Email.append(Customer[i][1])
    Phone_no.append(Customer[i][2])
    SSN.append(Customer[i][3])
    Address.append(Customer[i][4])
```
