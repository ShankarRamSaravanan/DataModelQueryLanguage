```python
import numpy as np

import random
    
    
random.randrange(100000,5000000,100000)

loan_amount_requested =[]
for i in range(5000):
    loan_amount_requested.append(random.randrange(100000,5000000,100000))

    
#loan_amount_requested 

loan_amount_sanctioned=[]
for i in loan_amount_requested:
    #print(np.random.uniform(0.72,1),)
    amount = i * round(np.random.uniform(0.72,1),2)
    loan_amount_sanctioned.append(int(amount))
    
#loan_amount_sanctioned   





loan_amount_disbursed=[]
for i in loan_amount_sanctioned:
    #print(np.random.uniform(0.72,1),)
    amount = i * round(np.random.uniform(0,1),2)
    loan_amount_disbursed.append(int(amount))
    
loan_amount_disbursed   
    
```

```python
Loan_id = np.random.randint(500051111111,500059999999,size=(5000), dtype=np.int64)

Loan_id = Loan_id.tolist()
Loan_id
```

```python

collateral= np.random.randint(3,6,size=(5000))


updated_Collateral=[]
for i,j in zip(loan_amount_sanctioned,collateral):
    updated_Collateral.append(i*j)
    
updated_Collateral
    
    

```
