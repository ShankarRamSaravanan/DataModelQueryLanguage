```python
import pandas as pd
employee = pd.read_csv("C:/ram/FALL/DMQL/project/EMployee.csv")
```

```python
employee
```

```python

Upd_branch_id=[]
for i in employee['Branch_id']:
    Upd_branch_id.append(i.replace("BOFA","BSA"))
    
    
Upd_emp_id=[]
for i in employee['Employee_Id']:
    Upd_emp_id.append(i.replace("BSA","E"))


    
employee=employee.drop(["Branch_name",'Employee_Id'],axis=1)
employee['Branch_id'] = Upd_branch_id


employee.insert(0, 'Employee_id', Upd_emp_id)
```

```python
employee.to_csv('Employee.csv')
```
